```python
# DataFrame
exam_data = {
    '수학': [90, 80, 70],
    '영어': [90.0, 70.0, 90.0],
    '성별': ['남', '여', '남'],
    '합격': [True, False, True],
}
exam_df = pd.DataFrame(exam_data)
print(exam_df)
```

```python
## Summary statistic
exam_df.shape       # 데이터프레임의 행과 열의 개수 반환
exam_df.columns     # 데이터프레임의 열 이름 반환
exam_df.dtypes      # 각 열의 데이터 타입 반환
exam_df.info()      # 데이터프레임의 전체적인 요약 정보 제공
```

```python
##
df1 = pd.DataFrame({'a': ['a', 'a1', 'a2', 'a3'],
                    'b': ['b', 'b1', 'b2', 'b3'],
                    'c': ['c', 'c1', 'c2', 'c3']},
                   index = [0, 1, 2, 3])

df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['c2', 'c3', 'c4', 'c5']},
                   index = [2, 3, 4, 5])

print('df1 = \n', df1)
print('df2 = \n', df2)
```

```python
## Concatenation # 데이터를 있는 것만 병합? 하겠다 할때 가장 편한것
# 데이터를 살릴때는 join = 'outer' 데이터가 없는 행렬을 잘라내려면 join = 'inner'
concat_row = pd.concat([df1, df2], join = 'inner', ignore_index=True)
print('pd.concat([df1, df2]) = \n', concat_row)
# 행을 이어 붙일때는 그냥 붙이면 같은 column끼리 붙지만 # 인덱스를 기준으로 열을 병합하려면 axis = 1
concat_col = pd.concat([df1, df2], axis = 1) 
print('pd.concat([df1, df2], axis = 1) = \n', concat_col)
# concat이나 join 은 인덱스로붙인다.
```

```python
## 
df1 = pd.read_excel("./dataset/stock price.xlsx")
df2 = pd.read_excel("./dataset/stock valuation.xlsx")
##
df1.set_index("id", inplace = True)
df2.set_index("id", inplace = True)

print("df1 = \n", df1)
print("df2 = \n", df2)
```

```python
df3 = df1.join(df2, lsuffix = "x", rsuffix = "y", how = 'inner')
# suffix는 접두사를 의미한다. 접두사 df1에는 x df2에는 y가 붙는다.
# suffixes = ("df1", "df2") 뭐더라
df4 = df2.join(df1, how = 'inner')
print("df3 = \n", df3)
print("df4 = \n", df4)
```

```python
## merge # concat이나 join과 달리 열로 붙일 수 있어 자유도가 높다.
# merge_inner = pd.merge(df1, df2, how = 'inner', on = 'id')
# print('merge_inner = \n', merge_inner)

# merge_outer = pd.merge(df1, df2, how = 'inner', on = 'id')
# print('merge_inner = \n', merge_inner)

##
merge_df1 = pd.merge(df1, df2, how = 'left', left_on = 'stock_name', right_on = 'name')
print("merge_df1 = \n", merge_df1)
```
