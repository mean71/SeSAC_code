# 데이터 분석 기초



## Data 의 종류
### 연속형 자료
- 등간척도 (Interval): 절대값 0 이 존재하지 않음, 곱셈법칙 적용이 안됨, 예)온도
- 비율척도 (Ratio): 절대값 0 이 존재, 예) 키, 몸무게, 혈압
### 범주형 자료
- 명목척도 (Nominal): 속성을 분류하는 척도, 속성간 순위가 없음, 예) 남/녀, 국어/영어/수학
- 순서척도 (Ordinal): 속성간에 순위개념이 존재하는 척도, 예) 상/중/하, 좋음/보통/나쁨

## 요약통계량 구하기

### seaborn 예제 자료

```python
import seaborn as sns
print(sns.get_dataset_names())
mpg_df = sns.load_dataset('mpg')
print("mpg_df = \n", mpg_df.head)
```

### 이산형 자료 통계량 구하기

```python
print(mpg_df['cylinders'].value_counts()) # value_counts() 해당 열에서 요소 고유값의 등장 빈도
print(mpg_df['origin'].value_counts())
```

### 연속형 자료 통계량 구하기

```python
print(mpg_df['mpg'].mean().round(3))   # mpg의 평균을 소수점 셋째 자리에서 반올림
print(mpg_df['mpg'].median())          # mpg의 중앙값(중위값)
print(mpg_df['mpg'].var().round(3))    # mpg의 분산을 소수점 셋째 자리에서 반올림
print(mpg_df['mpg'].std().round(3))    # mpg의 표준편차를 소수점 셋째 자리에서 반올림
print(mpg_df['mpg'].sem().round(3))    # mpg의 표준오차를 소수점 셋째 자리에서 반올림
print(mpg_df['mpg'].max())             # mpg의 최댓값
print(mpg_df['mpg'].min())             # mpg의 최솟값
```

```python
print(mpg_df[['mpg', 'weight']].mean()) # 'mpg'와 'weight'열의 평균값을 계산하여 Series 형태로 출력.
print(mpg_df[['mpg', 'acceleration']].var()) # 'mpg'와 'acceleration' 열의 분산을 계산하여 출력합니다.
print("*"*20)

print('quantile 25% =')
print(mpg_df[['mpg', 'acceleration']].quantile(0.25)) # 'mpg'와 'acceleration' 열의 25% 분위수 값을 계산하여 출력합니다.
print("*"*20)

print('quantile 75% = ')
print(mpg_df[['mpg', 'acceleration']].quantile(0.75)) # 'mpg'와 'acceleration' 열의 75% 분위수 값을 계산하여 출력합니다.
print("*"*20)

print('standard error =')
print(mpg_df[['mpg', 'acceleration']].sem()) # 'mpg'와 'acceleration' 열의 표준오차를 계산하여 출력합니다.
```

## Matplotlib 을 이용하여 산점도 그리기

```python
print('correlation coefficient = \n') # 'mpg'와 'weight' 열 간의 상관 계수를 계산하고 출력. 상관 계수는 두 변수 간의 선형 관계의 강도. # -1에서 1까지의 값으로 표현됩니다.
# 1은 완벽한 양의 선형 관계, -1은 완벽한 음의 선형 관계를 의미. 0에 가까운 값은 선형 관계가 없음
print(mpg_df[['mpg', 'weight']].corr()) # .corr() 상관계수를 계산하는 pd메서드

plt.scatter(mpg_df['weight'], mpg_df['mpg'], s = 20) # 's=20' :점 크기
# 산점도 그래프를 생성하여 'weight'와 'mpg' 간의 관계를 시각화.
plt.title("Weight vs MPG")  # 그래프의 제목
plt.xlabel('weight(kg)')  # x축 레이블 설정
plt.ylabel('mpg')  # y축 레이블 설정
plt.show()  # 그래프 출력
```
