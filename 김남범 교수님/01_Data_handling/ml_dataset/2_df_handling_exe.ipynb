```python
## 문제
df = pd.DataFrame({
    'hdz' : ['yes', 'no', 'no', 'no', 'yes'],
    'chestpain' : [True, False, False, True, False],
    'cholesterol' : [208, 282, 235, 277, 280],
    'sysbp' : [160, 140, 188, 162, 122]
}, index = np.arange(1, 6))
print(df)
################################### loc를 이용한 문제
print("loc를 이용한 문제")
print('문제 1. chestpain의 3번째와 5번째 자료를 loc index를 이용하여 추출하여라')
print("1: \n", df.loc[[2, 4], "chestpain"])
print("문제 2. chestpain, cholesterol, sysbp의 변수들의 2번째에서 ~ 4번째 자료를 loc index를 이용하여 추출하여라")
print("2: \n", df.loc[1:3, 'chestpain':'sysbp'])
print("문제 3. hdz 열 전체를 loc index를 이용하여 추출하여라")
print("3: \n", df.loc[:, "hdz"])
print("문제 4. 열의 순서를 바꾸어서 loc index를 이용하여 추출하여라")
print("4: \n", df.loc[:, ::-1])
################################### loc를 이용한 문제
print("문제 1. chestpain의 3번째와 5번째 자료를 iloc index를 이용하여 추출하여라")
print("1: \n", df.loc)
print("문제 2. chestpain, cholesterol, sysbp의 변수들의 2번째에서 ~ 4번째 자료를 iloc index를 이용하여 추출하여라")
print("2: \n", df.loc)
print("문제 3. hdz 열 전체를 iloc index를 이용하여 추출하여라")
print("3: \n", df.loc)
print("문제 4. 열의 순서를 바꾸어서 iloc index를 이용하여 추출하여라")
print("4: \n", df.loc)
```
