```python
## 연속형 자료
# mpg_df.head()
print(mpg_df["mpg"].mean(skipna = True).round(3)) # 평균 # 시리즈타입이라 axis는 없다
print(mpg_df["mpg"].median()) #
print(mpg_df["mpg"].var().round(3)) # 분산 varias? 데이터의 퍼짐의 정도
print(mpg_df["mpg"].std().round(3)) #
print(mpg_df["mpg"].sem().round(3)) # varias? 스탠다드 엘오민?데터가 아니라 평균의 어쩌고
print(mpg_df["mpg"].max())
print(mpg_df["mpg"].min())

## 이산형
print(mpg_df["cylinders"].value_counts())
print(mpg_df["origin"].value_counts())
```

```python
print(mpg_df[["mpg", "weight"]].mean(axis = 0).round(3))
print(mpg_df[["mpg", "weight"]].var(axis = 0).round(3))

print("*"*100)
print('quantile = 0.25')
print(mpg_df[["mpg", "weight"]].quantile(0.25))
```

```python
## correlation
print("상관계수 = \n")
print("correlation coefficient = \n")
print(mpg_df[["mpg", "weight"]].corr())

plt.scatter(mpg_df["weight"], mpg_df["mpg"])
plt.title("Weight vs MPG")
plt.xlabel("Weight (kg)")
plt.ylabel("MPG (mile/gallon)")

# plt.scatter(mpg_df["mpg"], mpg_df["weight"])

plt.show()
#사회과학에선 0.7 이면 최고의 상관성이지만 실험에선 최소가 0.7
# -0.831741 이면 자동차의 무게가 늘어나면 연비가 떨어지는 음의 상관성을 의미한다.
# 그림을 그리는데 기본적으로 (x축이 원인변수이다. 있어보이는 말로 설명?변수 또는 독립변수) (y축은 결과변수, 독립일때는 종속변수)
# 회계분석할때 인과성 분석할때 의미있다
# 인과성분석 상관성분석
# 영어점수 높은데 수학점수도 높은것처럼 인과성이 성립하지않지만 정확히는 모르겠지만 상관성이 있을때 상관성이 있다? 맞나?
```
