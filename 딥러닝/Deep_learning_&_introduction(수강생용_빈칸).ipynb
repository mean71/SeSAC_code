### ì˜¤ëŠ˜ì˜ ë³µìŠµìš© ë¬¸ì œ

1. l = list of list of .... ë¥¼ ë°›ì•„ì„œ, í•˜ëŠ” í•¨ìˆ˜ get_shapeë¥¼ ì‘ì„±í•˜ì„¸ìš”.
  - ì—¬ê¸°ì„œ torch.tensor(l).shape ê³¼ ê°™ì€ ê°’ì„ ê°€ì§€ëŠ” ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜
  - ë§Œì•½ torch.tensorê°€ ë¶ˆê°€ëŠ¥í•˜ë‹¤ë©´ Falseë¥¼ ë¦¬í„´
2. l = list of list of ..., r = list of list of ... ë‘ inputì„ ë°›ì•„ì„œ, ì•„ë˜ì™€ ê°™ì´ ë™ì‘í•˜ëŠ” í•¨ìˆ˜ broadcastingì„ ì‘ì„±í•˜ì„¸ìš”
  - ë¸Œë¡œë“œìºìŠ¤íŒ…ì´ ë  ë•Œ ê° l, rì´ ë°”ë€Œì–´ì•¼ í•˜ëŠ” í˜•íƒœë¥¼ ë¦¬í„´

````python
def get_shape(lst):
    res = []
    cur = lst

    while isinstance(cur, list):
        res.append(len(cur))
        cur = cur[0]

    return res

def get_shape(lst):
    if not isinstance(lst, list):
        return []
    else:
        shapes = []
        for elem in lst:
            shape = get_shape(elem)
            if shape not in shapes:
                shapes.append(shape)
        if len(shapes) == 1:
            return [len(lst)] + get_shape(lst[0])
        else:
            return False

def fill(l, r):
    """If len(l) > len(r), fill 1 to r's front, so that len(l) == len(r),
    If len(r) < len(l), do the opposite.
    """
    if len(l) > len(r):
        diff = len(l) - len(r)
        r = [1 for _ in range(diff)] + r
        return l, r
    elif len(l) < len(r):
        diff = len(r) - len(l)
        l = [1 for _ in range(diff)] + l
        return l, r
    return l, r

def expand_dimension(l, dim_idx, r_s):
    """
    l = [[1,2,3,]] (shape 1, 3)
    dim_idx = 0
    r_s = 4
    expand_dimension(l, 0, 4)
    >> [[1,2,3,], [1,2,3,], [1,2,3,], [1,2,3,]]

    l = [[[1,2,3]], [[1,2,3]], [[1,2,3]]] (shape 3, 1, 3 -> 3, 2, 3)
    l[0] = [[1,2,3]], l[1], l[2] (shape 2, 3)
    dim_idx = 1
    r_s = 2
    expand_dimension(l, 0, 2)
    >> [[[1,2,3], [1,2,3]], [[1,2,3], [1,2,3]], [[1,2,3], [1,2,3]]] (shape 3, 2, 3)

    l / shape 4, 3, 2, 1, 2 -> 4, 3, 2, 5, 2
    expand_dimension(l, 3, 5)
    l[0], l[1], l[2], l[3] / shape 3, 2, 1, 2 -> 3, 2, 5, 2
    expand_dimension(l[0], 2, 5)
    expand_dimension(l[1], 2, 5)
    expand_dimension(l[2], 2, 5)
    expand_dimension(l[3], 2, 5)
    """
    assert get_shape(l)[dim_idx] == 1, (get_shape(l), dim_idx)

    if dim_idx == 0:
        return [l[0] for _ in range(r_s)]
    else:
        return [expand_dimension(e, dim_idx - 1, r_s) for e in l]


def broadcasting(l, r):
    shape_l = get_shape(l)
    shape_r = get_shape(r)

    assert shape_l and shape_r

    # ì°¨ì›ì˜ ë§ì¶”ê¸°: ë‘ í…ì„œì˜ ì°¨ì›(Dimension) ìˆ˜ê°€ ë‹¤ë¥¼ ë•Œ,
    # ì°¨ì›ì´ ì‘ì€ í…ì„œì˜ ì•ìª½ì— 1ì„ ì¶”ê°€í•˜ì—¬ ì°¨ì›ì„ ë§ì¶¥ë‹ˆë‹¤.

    # (2, 3) / (4, 5, 2, 3) -> (1, 1, 2, 3) / (4, 5, 2, 3)

    l_is_bigger = False
    r_is_bigger = False
    diff = abs(len(shape_l) - len(shape_r))

    if len(shape_l) > len(shape_r):
        l_is_bigger = True
    elif len(shape_l) < len(shape_r):
        r_is_bigger = True

    shape_l, shape_r = fill(shape_l, shape_r)

    for _ in range(diff):
        if l_is_bigger:
            r = [r]  # r.shape: a1, a2, ... , an / [r].shape : 1, a1, a2, ... , an
        elif r_is_bigger:
            l = [l]

    assert shape_l == get_shape(l)
    assert shape_r == get_shape(r)

    # í¬ê¸° ë§ì¶”ê¸°: ê° ì°¨ì›ì—ì„œ í¬ê¸°ê°€ 1ì¸ í…ì„œëŠ”
    # í•´ë‹¹ ì°¨ì›ì˜ í¬ê¸°ë¥¼ í° í…ì„œì˜ í¬ê¸°ì— ë§ì¶° ëŠ˜ë¦´ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

    dim_idx = 0

    for l_s, r_s in zip(shape_l, shape_r):
        if l_s != r_s:
            if min(l_s, r_s) == 1:
                if l_s == 1: #
                    l = expand_dimension(l, dim_idx, r_s)
                else: # r_s == 1
                    r = expand_dimension(r, dim_idx, l_s)
            else:
                return False
        dim_idx += 1

    return l, r

l = [[[1,2,3]], [[1,2,3]], [[1,2,3]]]
print("get_shape(I) :", get_shape(l))

r = [[1,2,3,], [1,2,3,], [1,2,3,], [1,2,3,]]
print(get_shape(r))

# 3 1 3 / 4 3 -> 3 1 3 / 1 4 3 -> 3 1 3 / 3 4 3 -> 3 4 3 / 3 4 3
# r = [[[1,2,3,], [1,2,3,], [1,2,3,], [1,2,3,]]]
r_ans = [[
            [1,2,3,],
            [1,2,3,],
            [1,2,3,],
            [1,2,3,]
          ],
         [[1,2,3,], [1,2,3,], [1,2,3,], [1,2,3,]],
         [[1,2,3,], [1,2,3,], [1,2,3,], [1,2,3,]]]
l_ans = [[[1,2,3], [1,2,3], [1,2,3], [1,2,3]],
         [[1,2,3], [1,2,3], [1,2,3], [1,2,3]],
         [[1,2,3], [1,2,3], [1,2,3], [1,2,3]]]
l, r = broadcasting(l, r)
print(get_shape(l), get_shape(r))
print(l_ans == l, r_ans == r)
````

````python
#epochì€ ì´ ë°ì´í„°ë¥¼ ëª‡ë°”í€´ ëŒë ¤ì„œ í•™ìŠµì„ ì‹œí‚¬ ê²ƒì´ëƒ! ë¼ê³  ë³¼ ìˆ˜ ìˆê² ì£ ? ê¸°ì–µë‚˜ì‹œë‚˜ìš”?
for epoch in range(epochs):

    #ìš°ë¦¬ëŠ” yì„ ì˜ˆì¸¡í•˜ê³  ì‹¶ìŠµë‹ˆë‹¤. yì— ì˜í–¥ì„ ì£¼ëŠ” ë³€ìˆ˜ëŠ” Xí•˜ë‚˜ ë¼ê³  ìƒê°í•˜ê³  ê°„ë‹¨í•œ ì„ í˜•ì‹ì„ ë§Œë“¤ì–´ ë³´ì•˜ìŠµë‹ˆë‹¤!
    y_pred = X * w + b # y_pred: 100, 1


    # Compute and print loss
    # loss í•¨ìˆ˜ë¡œ ìš°ë¦¬ëŠ” MSE í‰ê· ì œê³±ì˜¤ì°¨ì˜ ì‹ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.
    # ëª¨ë“ˆì„ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ìˆê² ì§€ë§Œ ìˆ˜í•™ì‹ìœ¼ë¡œ ì‘ì„±í•˜ë©´ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.
    # ì˜ˆìƒí•œ ê°’ - ì‹¤ì œê°’ì˜ ì œê³±ì˜ í‰ê· 
    loss = torch.mean(y_pred**2) # 100, 1


    # 100 ì—í­ë‹¹ ìš°ë¦¬ëŠ” ì§€ê¸ˆ í•™ìŠµì´ ì–´ëŠì •ë„ ë˜ê³  ìˆëŠ”ì§€ ìˆ˜ì¹˜ë¡œ ë³´ê³  ì‹¶ì–´ìš”!
    # ê·¸ë˜ì„œ if%100ì´ 0ì¼ë•Œ print í•´ë³´ëŠ” ì½”ë“œë¥¼ ì§œë³´ì•˜ìŠµë‹ˆë‹¤.
    if epoch % 100 == 0:
        print(f'Epoch {epoch}: w = {w.item():.4f}, b = {b.item():.4f}, loss = {loss.item():.4f}')
    # Forward pass: compute predicted y
    # 100, 1 / 1 -> 100, 1 / 1, 1 -> 100, 1 / 100, 1


    # Backward pass: compute gradients
    # ì—­ì „íŒŒë¥¼ í†µí•´ ê¸°ìš¸ê¸°ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
    # loss ëŠ” ìš°ë¦¬ê°€ ìœ„ì— ì •ì˜í–ˆë˜ 'torch.mean((y_pred - y) ** 2)' ì´ì—ˆìŠµë‹ˆë‹¤!
    # ì—¬ê¸°ì„œ ê¸°ìš¸ê¸°ë¥¼ ê³„ì‚°í•˜ê²Œ ë˜ë©´ ë…ë¦½ë³€ìˆ˜(X) ì— ì˜í–¥ì„ ì£¼ì—ˆë˜ w, b ì— ëŒ€í•´ì„œ ê¸°ìš¸ê¸° ê°’ì´ ë‚˜ì˜¤ê²Œ ë©ë‹ˆë‹¤.
    loss.backward()


    # Update parameters using gradient descent
    # ì ê¹ ! ì—¬ê¸°ì„œ ì™œ - ë¥¼ í• ê¹Œìš”?
    # ìœ„ ê·¸ë¦¼ì—ì„œ ìš°ë¦¬ëŠ” ê¸°ìš¸ê¸°ê°€ ìŒìˆ˜ì´ë©´ \ ì–‘ìˆ˜ì´ë©´ / ëª¨ì–‘ì„ ë„ëŠ” ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆì—ˆìŠµë‹ˆë‹¤.
    # ìš°ë¦¬ëŠ” ê²½ì‚¬í•˜ê°•ë²• ê·¸ë˜í”„ì—ì„œ ë³¼ ìˆ˜ ìˆë“  ìµœì €ì ì„ ì°¾ì•„ê°€ì•¼ í•˜ëŠ”ë°
    # ê·¸ëŸ¼ ê¸°ìš¸ê¸°ê°€ ìŒìˆ˜ì´ë©´ ì˜¤ë¥¸ìª½ ì–‘ìˆ˜ì´ë©´ ì™¼ìª½ìœ¼ë¡œ ê°€ì•¼í•˜ê² ì£ ?
    # ì¦‰ ìŒìˆ˜ì´ë©´ ê·¸ë˜í”„ìƒ Xì¶• ê¸°ì¤€ìœ¼ë¡œ + ì–‘ìˆ˜ì´ë©´ Xì¶• ê¸°ì¤€ìœ¼ë¡œ - ë¥¼ í•´ì•¼í•˜ëŠ” ê²ƒ ì…ë‹ˆë‹¤.ì´ê±¸ ì–´ë–»ê²Œ í• ê¹Œìš”?
    # ë°”ë¡œ -ì—°ì‚°ì„ í•˜ë©´ ë©ë‹ˆë‹¤. -ì— -ë¥¼ í•˜ë©´ +ê°€ ë˜ëŠ” ê²ƒì€ ì•„ì‹œì£ ?
    # ê·¸ë˜ì„œ ê¸°ìš¸ê¸°ê°€ ìŒìˆ˜ì¼ ê²½ìš° ë¹¼ì£¼ë©´ +ê°€ ë˜ì–´ ì˜¤ë¥¸ìª½ìœ¼ë¡œ ì´ë™í•˜ê³  ê¸°ìš¸ê¸°ê°€ ì–‘ìˆ˜ì¼ë•Œ ë¹¼ì£¼ë©´ ì™¼ìª½ìœ¼ë¡œ ì´ë™í•˜ê²Œ ë˜ëŠ” ê²ƒì´ì£ !
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

    # Zero gradients after updating
    # ê·¸ë¦¬ê³  ì´í›„ í•™ìŠµì„ ìœ„í•´ ê¸°ìš¸ê¸°ë¥¼ ì´ˆê¸°í™” ì‹œì¼œì¤ë‹ˆë‹¤.
    
    
    w.grad.zero_()
    b.grad.zero_()
````

````python
# write your code here
import random


# ì ì´ë¶€ë¶„ì€ ë°ì´í„°ë¥¼ ì¸ê³µì ìœ¼ë¡œ í˜•ì„±í•˜ëŠ” ì½”ë“œì…ë‹ˆë‹¤. ì´ ë¶€ë¶„ì€ ë‹¹ì¥! ì´í•´ê°€ ë˜ì§€ ì•Šìœ¼ì‹ ë‹¤ë©´
# ë„˜ì–´ê°€ì£¼ì„¸ìš” ì•„ë˜ í•¨ìˆ˜ë¥¼ í†µí•´ ëª¨ë¸ì— ë°ì´í„°ë¥¼ ì „ë‹¬í•´ì¤„ ê²ƒ ! ì´ë¼ëŠ” ì  ë§Œ ê¸°ì–µí•´ì£¼ì„¸ìš”!
def generate_artificial_data(true_w, true_b, n_data,
                                x_amplitude = 10,
                                noise_amplitude = 0.2):
    """Generate data from y = true_w * x + true_b
    """
    X = torch.randn(n_data, 1) * x_amplitude
    y = true_w * X + true_b + torch.randn(n_data, 1) * noise_amplitude

    return X, y

# ì yëŠ” ìš°ë¦¬ê°€ ì˜ˆì¸¡í•  ê°’ XëŠ” yì— ëŒ€í•´ ì˜í–¥ì„ ì£¼ëŠ” ì–´ë– í•œ ë³€ìˆ˜ì…ë‹ˆë‹¤!
def linear_regression(data, learning_rate, epochs, quiet = False):
    #data(ìœ„ ì •ì˜í•œ 'generate_artificial_data' í•¨ìˆ˜ë¥¼ ì—¬ê¸°ì— ì‚¬ìš©í•  ì˜ˆì •) ë¥¼ í†µí•´ X,yë¥¼ ë°›ì•„ì™”ìŠµë‹ˆë‹¤.
    X, y = data

    #ì‹¤ì œë¡œ ë”¥ëŸ¬ë‹ì„ ëŒë¦´ë•Œë„ ì´ˆê¸° ê°€ì¤‘ì¹˜ì™€ í¸í–¥(weight, bias) ê°’ì€ ëœë¤í•˜ê²Œ ì°í™ë‹ˆë‹¤.
    #ëœë¤í•˜ê²Œ ê°€ì¤‘ì¹˜ì™€ í¸í–¥ì„ ìƒì„±í•©ë‹ˆë‹¤.
    #ì•„ë˜ ë¬¼ìŒí‘œì— ê¸°ìš¸ê¸°ë¥¼ ê¸°ë¡í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì‹œì˜¤
    w = torch.randn(1, 1, requires_grad=True)
    b = torch.randn(1, requires_grad=True)

    # ì—í­ì€ ì•„ê¹Œ ë§í–ˆë“¯ ë°ì´í„°ë¥¼ epochíšŸìˆ˜ ë§Œí¼ ëŒë ¤ì„œ í•™ìŠµí•œë‹¤ê³  ë§ì”€ë“œë ¸ì£ ?
    for epoch in range(epochs):

        # Forward pass: compute predicted y
        # 100, 1 / 1 -> 100, 1 / 1, 1 -> 100, 1 / 100, 1
        # ìœ„ ì˜ˆì œì™€ ë˜‘ê°™ì€ ì„ í˜•ì‹ì…ë‹ˆë‹¤.
        y_pred = X * w + b # y_pred: 100, 1

        # Compute and print loss
        # ë˜‘ê°™ì€ ì†ì‹¤í•¨ìˆ˜ì¸ MSE ì…ë‹ˆë‹¤.
        loss = torch.mean((y_pred - y) ** 2) # 100, 1

# ì•„ë˜ ì½”ë“œëŠ” ìœ„ì— ì„¤ëª…ë“œë ¸ë˜ ì½”ë“œì™€ ê°™ì€ ì½”ë“œë“¤ì…ë‹ˆë‹¤. ì—¬ëŸ¬ë¶„ë“¤ì´ ì–´ë–¤ ì½”ë“œì˜€ëŠ”ì§€ ëŒ€ë‹µí•´ë³¼ê¹Œìš”?
        if epoch % 100 == 0 and not quiet:
            print(f'Epoch {epoch}: w = {w.item():.4f}, b = {b.item():.4f}, loss = {loss.item():.4f}')
        # Backward pass: compute gradients
        loss.backward()

        # Update parameters using gradient descent
        with torch.no_grad():
            w -= learning_rate * w.grad
            b -= learning_rate * b.grad

        # Zero gradients after updating
        w.grad.zero_()
        b.grad.zero_()

    return w, b

data = generate_artificial_data(2.0, 1.0, 1000)
linear_regression(data, 0.001, 1000)

#---------------------------------------------

#ì—¬ê¸°ì„œ ë¶€í„´ ìš°ë¦¬ê°€ í•™ìŠµí•œ ëª¨ë¸ì„ í‰ê°€(evaluate) í•´ë³´ëŠ”  ì½”ë“œì…ë‹ˆë‹¤.
#ì‹¤ì œ í˜„ì—…ì—ì„œëŠ” ì‹¤ì œ ê°€ì¤‘ì¹˜ì™€ ì‹¤ì œ í¸í–¥ì„ ì•Œ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. í•˜ì§€ë§Œ ì‹¤ì œ ê°€ì¤‘ì¹˜ì™€ í¸í–¥ê³¼ ìš°ë¦¬ê°€ í•™ìŠµí•œ
#ëª¨ë¸ì´ ì˜ˆì¸¡í•˜ëŠ” ê°’ì´ ì–¼ë§ˆë‚˜ ê°€ê¹Œì›Œì¡ŒëŠ” ì§€ í…ŒìŠ¤íŠ¸ í•´ë³´ê¸° ìœ„í•´ ë‹¤ìŒê³¼ ê°™ì€ í•¨ìˆ˜ë¥¼ ì‘ì„±í•´ë³´ì•˜ìŠµë‹ˆë‹¤.
def evaluate_model(true_w, true_b, w, b):
    return (true_w - w) ** 2 / true_w ** 2 + (true_b - b) ** 2 / true_b ** 2


def find_learning_rate():
    #ì‹¤ì œ í¸í–¥ê³¼ ê°€ì¤‘ì¹˜ë¥¼ ëœë¤í•˜ê²Œ ìƒì„±
    true_w = random.gauss(0, 10)
    true_b = random.gauss(0, 10)

    data = generate_artificial_data(true_w, true_b, 1000)

    for learning_rate in [0.0001 * (i+1) for i in range(10)]:
        w, b = linear_regression(data, learning_rate, 1000, quiet = True)
        score = evaluate_model(true_w, true_b, w, b)
        print(score, learning_rate)

find_learning_rate()
````
